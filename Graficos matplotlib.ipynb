```python
# Anova

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import f_oneway

# Datos de ejemplo
grupo1 = np.random.normal(20, 5, 100)
grupo2 = np.random.normal(25, 5, 100)
grupo3 = np.random.normal(30, 5, 100)

# realizar el Anova
f_statistics, p_value = f_oneway(grupo1, grupo2, grupo3)

print("Estadistica F:", f_statistics)
print("Valor P:", p_value)

# grafico
plt.boxplot([grupo1, grupo2, grupo3], labels= ['Grupo 1', 'Grupo 2', 'Grupo3'])
plt.ylabel('Valores')
plt.title('Boxplot')
plt.show()
```
